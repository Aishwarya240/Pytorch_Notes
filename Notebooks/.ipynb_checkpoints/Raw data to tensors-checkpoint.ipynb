## Working with kaggle dataset

```python
import pandas as pd
```

```python
data = pd.read_csv("Datasets/heart.csv")
data.head()
```

```python
data.info()
```

```python
X = data.drop(["HeartDisease"], axis=1)
X.head()
```

```python
y = data["HeartDisease"]
y.head()
```

Convert string data types columns to int data types

```python
# Using scikit-learn label encoder
from sklearn.preprocessing import LabelEncoder

a = ["M", "F", "M"]

le = LabelEncoder()

v = le.fit_transform(a)
v
```

```python
le.inverse_transform(v)
```

```python
# Example column from string datatype to int data type
a = X["Sex"]
f = le.fit_transform(a)
f
```

```python
le = LabelEncoder()
col = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]

X[col] = X[col].apply(le.fit_transform)

X
```

```python
X.info()
```

```python
X_col = X.columns
X_col
```

```python
for X_col in X:
    print(X[X_col])
```

```python
X.values
```

```python
# Pandas Dataframe (numpy_arrays) to torch tensors
X_tensors = torch.from_numpy(X.values).to(device="cuda", dtype=torch.float16)
X_tensors
```

```python
y_tensors = torch.tensor(y).to(device="cuda", dtype=torch.float16)
y_tensors
```
