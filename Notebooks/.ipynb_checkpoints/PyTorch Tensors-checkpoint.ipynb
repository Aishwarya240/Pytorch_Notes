Tensor Operations

```python
a = torch.tensor([1, 2, 3, 4, 5],device=torch.device("cuda"))
a
```

```python
torch.sum(a).to(device=torch.device("cuda"), dtype=torch.int64)
```

```python
a = torch.tensor([1, 2, 3, 4, 5], device=torch.device("cuda"))
b = torch.tensor([1, 2, 3, 4, 5], device=torch.device("cuda"))

torch.sub(a, b).to(device=torch.device("cuda"))
```

```python
# Multiplication
torch.mul(a, b).to(device=torch.device("cuda"))
```

```python
a = torch.rand(3, device=torch.device("cuda"))
b = torch.rand(3, device=torch.device("cuda"))

```

```python
torch.mul(a, b).to(device=torch.device("cuda"))
```

### Comparison of CPU and GPU

```python
a = torch.rand(100000000, device=torch.device("cuda"))
```

```python
a
```

```python
a.shape
```

```python
import time
```

```python
s = time.time()
a.mul(a).to(device=torch.device("cpu"))
e = time.time() - s

print("CPU execution time: ", e)
```

```python
s = time.time()
a.mul(a).to(device=torch.device("cuda"))
e = time.time() - s

print("GPU execution time: ", e)
```
