## Student Marks Prediction using Regression Model

```python
import pandas as pd
```

```python
data = pd.read_csv('Datasets/Student_Marks.csv')
```

```python
data.head()
```

```python
X = data.drop(['Marks'], axis=1)
y = data['Marks']
```

```python
X.head()
```

```python
y.head()
```

- Now convert `X` and `y` into tensors

```python
import torch
```

```python
X_tensor = torch.tensor(X.values, dtype=torch.float)
y_tensor = torch.tensor(y.values, dtype=torch.float)
```

```python
X_tensor[:5]
```

```python
y_tensor[:5]
```

- train test split using sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

- Model training - Regression Model 

```python
import torch.nn as nn
```

```python
class Model(nn.Module):
    def __init__(self, input_dim=2, hidden_1=10, hidden_2=20, hidden_3=10, output_dim=1):
        super(Model, self).__init__()

        
        # layers    

        self.fc1 = nn.Linear(input_dim, hidden_1) #input layer

        # hidden state - start 
        self.fc2 = nn.Linear(hidden_1, hidden_2)
        self.relu1 = nn.ReLU()
        self.do1 = nn.Dropout(0.2)
        

        self.fc3 = nn.Linear(hidden_2, hidden_3)
        self.relu2 = nn.ReLU()
        self.do2 = nn.Dropout(0.1)

        self.fc4 = nn.Linear(hidden_3, output_dim)

    def forward(self, x):
        x = self.fc1(x)

        x = self.fc2(x)
        x = self.relu1(x)
        x = self.do1(x)
        

        x = self.fc3(x)
        x = self.relu2(x)
        x = self.do2(x)
        
        x = self.fc4(x)

        return x
```

```python
model = Model()
model 
```

```python
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
X_train.shape
```

```python
epoches = 101 

for epoch in range(epoches):
    model.train()

    out = model(X_train)
    loss = loss_function(out, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print("Epoch: ", epoch, "Loss: ", loss.item())
```

- Lets predict

```python
My_Data = torch.tensor([[1, 81.5]], dtype=torch.float)
```

```python
model.eval()
with torch.no_grad():
    out = model(My_Data)
```
