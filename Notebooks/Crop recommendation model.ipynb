# Model training

```python
import torch.nn as nn
```

```python
class CropRecommendationModel(nn.Module):
    def __init__(self):
        super(CropRecommendationModel, self).__init__()

        # layers
        self.fc1 = nn.Linear(7, 14)
        # ------ hidden layer
        self.fc2 = nn.Linear(14, 28)
        self.fc3 = nn.Linear(28, 56)
        self.fc4 = nn.Linear(56, 34)
        # -------
        self.fc5 = nn.Linear(34, 22) # output layer

    def forward(self, x):
        x = self.fc1(x) # input layer

        # hidden layer
        x = self.fc2(x)
        x = self.fc3(x)
        x = self.fc4(x)

        x = self.fc5(x) # output layer

    def forward(self, x):
        x = self.fc1(x) # input layer 
        
        # Hidden layer
        x = self.fc2(x)
        x = self.fc3(x)
        x = self.fc4(x)

        x = self.fc5(x) # output layer

        return x
```

```python
model = CropRecommendationModel()
```

```python
model
```

```python
loss_function = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training loop 
epoches = 100

for epoch in range(epoches):
    model.train()

    out = model(X_tensors)
    loss = loss_function(out, y_tensors.long().squeeze())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0 :
        print("Epoch: ", epoch, "Loss value: ", loss.item())
        
```

```python
y_pred = model(X_test).argmax(dim=1)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

### Using ReLU activation function


```python
class CropRecommendationModelReLU(nn.Module):
    def __init__(self):
        super(CropRecommendationModelReLU, self).__init__()

        # layers
        self.fc1 = nn.Linear(7, 14)
        self.relu1 = nn.ReLU()
        
        # ------ hidden layer
        self.fc2 = nn.Linear(14, 28)
        self.relu2 = nn.ReLU()
        self.do1 = nn.Dropout(0.2)
        
        self.fc3 = nn.Linear(28, 56)
        self.relu3 = nn.ReLU()
        self.do2 = nn.Dropout(0.2)
        
        self.fc4 = nn.Linear(56, 34)
        self.relu4 = nn.ReLU()
        self.do3 = nn.Dropout(0.1)
        # -------
        self.fc5 = nn.Linear(34, 22) # output layer

    def forward(self, x):
        x = self.fc1(x) # input layer
        x = self.relu1(x)
        

        # hidden layer
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.do1(x)
        
        x = self.fc3(x)
        x = self.relu3(x)
        x = self.do2(x)
        
        x = self.fc4(x)
        x = self.relu4(x)
        x = self.do3(x)

        x = self.fc5(x) # output layer

        return x
```
