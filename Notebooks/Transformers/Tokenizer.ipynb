# Tokenizer

### Tokenization using BertTokenization

```python
from transformers import BertTokenizer 
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
tokenizer
```

```python
text = "Sakib is a Yedzhava"
```

```python
tokens = tokenizer.tokenize(text)
tokens
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
token_ids
```

```python
# tokenizataion in one shot
tokenize_text = tokenizer(text)
tokenize_text 
```

```python
sentence = ["I am very cute", "I love sunflower"]

tokenized_sentence = tokenizer(sentence, padding=True, truncation=True, max_length=512, return_tensors='pt')
tokenized_sentence
```

```python
# ids to tokens 

my_text = "Hello this is me"

tokens = tokenizer.tokenize(my_text)

print("Tokens: ", tokens)

token_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Token Ids: ", token_ids)

ids_to_token = tokenizer.convert_ids_to_tokens(token_ids)
print("Ids to tokens: ", ids_to_token)

decode_sentence = tokenizer.decode(token_ids)
print("Decoded sentence: ", decode_sentence)
```
