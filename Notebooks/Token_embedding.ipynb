```python
import torch
```

```python
from transformers import BertTokenizer, BertModel
```

- There are two parts in transformers
    - Tokenizer
    - Encoder Model

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
text = "Hello world apple is a fruit"
```

```python
tokenizer(text)
```

```python
sentence = ["This is a mat!", "My cat sits on the mat", "She sleeps on the mat"]
```

```python
tokenizer(sentence)["input_ids"]
```

```python
tokenize_text = tokenizer(text)
```

```python
tokenize_text
```

```python
tokenize_text["input_ids"]
```

```python
tokenize_text = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=128)
tokenize_text["input_ids"]
```

```python
type(tokenize_text["input_ids"])
```

```python
model(**tokenize_text)
```

- Our encoding model expects input in PyTorch tensors.

```python
text = ["Hello world apple is a cat", "how are you?"]
```

```python
tokenized_text = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=128)
tokenized_text["input_ids"]
```

```python
tokenized_text
```

```python
type(tokenized_text["input_ids"])
```

```python
# **tokenized_text for token encoding
model(**tokenized_text).keys()
```

```python
with torch.no_grad():
    out = model(**tokenized_text)

out["last_hidden_state"]
```

```python
out["last_hidden_state"].shape
```

```python
new_out = out["last_hidden_state"][:, 0, :]
new_out
```
